```python
import os
from io import StringIO
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({'font.size': 12})
```

```python
ma_param = 5
```

# Train Bad0.5

```python
train_bad05_s = pd.read_csv("csvs/run_dn_s_st3d_1_logs-tag-bad_0_5.csv", sep=",")
train_bad05_c = pd.read_csv("csvs/run_dn_c_st3d_1_logs-tag-bad_0_5.csv", sep=",")
train_bad05_cs = pd.read_csv("csvs/run_dn_cs_st3d_3_logs-tag-bad_0_5.csv", sep=",")
```

```python
fig, ax = plt.subplots(figsize=(8, 6), linewidth=2)

ax.set(ylim=(20, 100), xlim=(-5, 105))
ax.set(xlabel='Epochs', ylabel='Bad 0.5 (%)')
ax.set_yticks(range(20,100,10))
ax.grid(axis='y')

ax.plot(
    list(range(0,100)),
    train_bad05_s.iloc[:, 2].rolling(3).mean().tolist(),
    label='Type S'
)
ax.plot(
    list(range(0,100)),
    train_bad05_c.iloc[:, 2].rolling(3).mean().tolist(),
    label='Type C'
)
ax.plot(
    list(range(0,100)),
    train_bad05_cs.iloc[:, 2].rolling(3).mean().tolist(),
    label='Type CS'
)

ax.legend()
fig.savefig("train-bad-05.png", dpi=300)
plt.show()
```
