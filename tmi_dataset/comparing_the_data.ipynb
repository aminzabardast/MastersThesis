```python
def boxplot(data, title=None, xlabels=None, ylabel=None, showfliers=True):
    fig, ax = plt.subplots(figsize=(8, 6), linewidth=5)
    ax.set_title(title)
    ax.set_ylabel(ylabel)
    ax.boxplot(data, showfliers=showfliers, labels=xlabels)
    plt.show()
```

```python
def barplot(data_1, data_2, title=None, xlabels=None, ylabel=None, colorlabels=None):
    n_groups = len(data_1)
    
    fig, ax = plt.subplots(figsize=(8, 6))

    index = np.arange(n_groups)
    bar_width = 0.35

    opacity = 1

    rects1 = ax.bar(index, data_1, bar_width,
                    alpha=opacity, color='#ea8893',
                    label=colorlabels[0])

    rects2 = ax.bar(index + bar_width, data_2, bar_width,
                    alpha=opacity, color='#730f13',
                    label=colorlabels[1])

    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.set_xticks(index + bar_width / 2)
    ax.set_xticklabels(xlabels)
    ax.legend()

    fig.tight_layout()
    plt.show()
```
